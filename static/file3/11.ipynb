```python
# apply函数
from datetime import datetime
data= pd.read_csv("df_merge_data_filter.csv")
data['Timestamp'] = data['Timestamp'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d %H:%M:%S'))
data['日期'] = data['Timestamp'].apply(lambda x: x.strftime('%Y-%m-%d'))
data['日期']
```

```python
# sort_values: 排序
data.sort_values(by='load(%)', ascending=False)
```

```python
# groupby：分组函数
data_group = data.groupby(by='日期').mean()
data_group[['AA yieldCoefficient(%mol)','PP conversionRate(%mol)']]
```

```python
# merge：左右拼接
a = data[['Timestamp','日期','AA yieldCoefficient(%mol)']]
b = data_group[['AA yieldCoefficient(%mol)']].reset_index()
pd.merge(a, b, how='left', left_on='日期', right_on='日期',suffixes=('_a', '_b'))
```

```python
# concat：上下拼接
pd.concat([a,b])
```
